```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly
```

```python
data_path = r'C:\Users\Abhishek\Desktop\Covid Analysis\Consolidated Data\complete.csv'
```

```python
data = pd.read_csv(data_path)
```

```python
data.tail(5)
```

```python
# Select the particular date for which we want to create Nightingle Rose Diagram
df = data[data['Date']=='16-05-2020']
# Remove India from the State List
df= df[df['Name of State / UT']!= 'India']
#Filter the dataframe with only the 4 required columns
df=df[['Name of State / UT','Total Confirmed cases','Cured/Discharged/Migrated','Death']]
#Select the 10 highest states of Confirmed Cases
df = df.nlargest(10,['Total Confirmed cases'])
#Sort according to State Names
df.sort_values(by=['Name of State / UT'],inplace=True)
```

```python
df
```

```python
fig = go.Figure()

fig.add_trace(go.Barpolar(
    r = list(df['Total Confirmed cases']),
    theta=list(df['Name of State / UT']),
    name = 'Confirmed',
    marker_color='rgb(46,109,255)',
    marker_line_color="black",
    hoverinfo = ['all'],
    opacity=0.7   
))


fig.add_trace(go.Barpolar(
    r = list(df['Cured/Discharged/Migrated']),
    theta=list(df['Name of State / UT']),
    name = 'Cured',
    marker_color='rgb(139,222,52)',
    marker_line_color="black",
    hoverinfo = ['all'],
    opacity=0.7
))

fig.add_trace(go.Barpolar(    
    r = list(df['Death']),
    theta=list(df['Name of State / UT']),
    name = 'Deceased',
    marker_color='rgb(245,94,21)', 
    marker_line_color="black",
    hoverinfo = ['all'],
    opacity=0.7
))


fig.update_layout(
    title='Covid Stats Analysis for Top 10 States in India',
    font_size=12,
    legend_font_size=15,
    polar_angularaxis_rotation=90,
    width=900,
    height=900,
    
    polar = dict(
              bgcolor = "rgb(223, 223, 223)",
              angularaxis = 
                dict(
                    linewidth = 3,
                    showline=True,
                    linecolor='black'
                    ),
          radialaxis = 
                dict(        
                    showline = True,
                    linewidth = 2,
                    gridcolor = "white",
                    gridwidth = 2,
                    )
                ),

            )
#fig.show()
plotly.offline.plot(fig, filename='NRD.html')
```
